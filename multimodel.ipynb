```python
# Customized Preprocessing
stop_words = set(stopwords.words('english'))
nlp = spacy.load("en_core_web_sm")
```

```python
def text_cleaning(text):
    # Remove special characters, digits, and convert to lowercase
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    cleaned_text = cleaned_text.lower()  # Convert text to lowercase
    return cleaned_text
```

```python
def select_keywords(text):
    doc = nlp(text)
    relevant_words = [token.text for token in doc if token.pos_ not in ['DET', 'ADP', 'CCONJ', 'PRON', 'PUNCT']]
    return ' '.join(relevant_words)
```

```python
# Content Selection & Preprocessing for the subset
for index, row in df1.iterrows():
    cleaned_text = text_cleaning(row['data'])
    selected_text = select_keywords(cleaned_text)
    df1.at[index, 'selected_text'] = selected_text
```

```python

# Fine-tuning the Summarization Model
summarizer = pipeline("summarization")
generated_summaries = []
```

```python
# Define a function to generate summaries for a single row's text
def generate_summary_for_row(text):
    max_chunk_length = 500  # Define your maximum chunk length
   # summarizer = pipeline("summarization")  # Your summarization model

    # Splitting text into segments of maximum sequence length
    segments = [text[i:i + max_chunk_length] for i in range(0, len(text), max_chunk_length)]

    chunk_summaries = []
    for segment in segments:
        # Perform summarization for each segment
        summary = summarizer(segment, max_length=20, min_length=5, do_sample=False)
        chunk_summaries.append(summary[0]['summary_text'])

    # Combine chunk summaries into a single summary for the entire text
    full_summary = ' '.join(chunk_summaries)
    return full_summary

# Loop through each row in the DataFrame
for index, row in df1.iterrows():
    input_text = row['selected_text']

    # Generate summaries for the current row's text
    generated_summary = generate_summary_for_row(input_text)

    # Add the generated summary to the DataFrame as a new column or store it as needed
    df1.at[index, 'generated_summary'] = generated_summary
```

```python
from rouge_score import rouge_scorer

# Initialize the ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Lists to store individual scores
rouge_r_scores = []
rouge_p_scores = []
rouge_f_scores = []

# Calculate ROUGE scores for each row in df1
for index, row in df1.iterrows():
    reference_summary = row['summary1']
    generated_summary = row['generated_summary']

    # Calculate ROUGE scores
    scores = scorer.score(reference_summary, generated_summary)

    # Append individual scores to lists
    rouge_r_scores.append(scores['rouge1'].recall)
    rouge_p_scores.append(scores['rouge1'].precision)
    rouge_f_scores.append(scores['rouge1'].fmeasure)

# Find the maximum scores among ROUGE-R, ROUGE-P, and ROUGE-F for each row
max_scores = [max(scores) for scores in zip(rouge_r_scores, rouge_p_scores, rouge_f_scores)]

# Calculate the average of the maximum scores
average_highest_score = sum(max_scores) / len(max_scores)

# Print the average highest score
print("Average of highest scores among ROUGE-R, ROUGE-P, and ROUGE-F:", average_highest_score)
```
